## post Operation 
- create a new object 
-  parameters sent via query string as weel as request body 
- requires an application framework eg . curl or requests 

Creates a new “user” on the API 

```python
import requests

url = "https://reqres.in/api/users"

headers = {
    "x-api-key": "reqres-free-v1",
    "Content-Type": "application/json"
}

data= {
    "name" : "ujjwal",
    "job": "Machine Learing Enginner"
}

response= requests.post(url, headers= headers , json = data)
```
